```python
#FUNCTION FOR ADDITION
def add(a,b):
    return a+b

#FUNCTION FOR SUBSTRACT    
def substract(a,b):
    return a-b

#FUNCTION FOR MULTIPLY    
def multiply(a,b):
    return a*b

#FUNCTION FOR DIVISION
def division(a,b):
    if b!=0:
       return a/b
    else:
        return "cannot divide by zero"

        
#EXAMPLE USAGE
print("add:",add(10,5))
print("substract:",substract(10,5))
print("multiply:",multiply(10,5))
print("division:",division(10,0))
```
