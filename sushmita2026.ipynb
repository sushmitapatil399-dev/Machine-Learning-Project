```python
#Example 2 of sample dataset
data={
    'Date':pd.date_range(start='2025-10-01',periods=10,freq='D'),
    'Product':['A','B','C','A','B','C','A','B','C','A'],
    'Units_sold':[10,15,8,12,18,10,14,9,20,16],
    'Unit_price':[50,40,60,50,40,60,50,60,40,50]
}

df=pd.DataFrame(data)
print(df)
print("____________________________________")
df['total_sales']=df['Units_sold']*df['Unit_price']
print(df)

#total revenue
total_revenue=df['total_sales'].sum()

#average units sold per product
avg_units_sold=df.groupby('Product')['Units_sold'].mean()

#highest selling product
top_product=df.groupby('Product')['Units_sold'].sum().idxmax()

print(f"Total Revenue:${total_revenue}")
print("\nAverage Units Sold per Product:\n{avg_units_sold}")
print(f"\nTop Selling Product:{top_product}")
```

```python
#example for linear regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#sample data (year of experience vs salary)
X=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([20000,25000,30000,35000,40000])

#create a linear regression model
model=LinearRegression()

#train model
model.fit(X,y)

#predict values
y_pred=model.predict(X)

#plot actual data points
plt.scatter(X,y)

#plot regression line
plt.plot(X,y_pred)

#labels
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("LinearRegression Example")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

#study hours(input)
X=np.array([1,2,3,4,5]).reshape(-1,1)

#result(0=fail,1=pass)
y=np.array([0,0,1,1,1])

#create model
model=LogisticRegression()
model.fit(X,y)

#create smooth curve
X_test=np.linspace(0,6,100).reshape(-1,1)
y_prob=model.predict_proba(X_test)[:,1]

#plot logistic curve
plt.plot(X_test,y_prob)

#labels
plt.xlabel("study hours")
plt.ylabel("probability of passing")
plt.title("Logistic Regression Example")
plt.show()

```

```python
#example support vector vs svm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

#step1:student data
#[study hours,attendance]
X=np.array([
    [1,2],
    [2,3],
    [3,3],
    [5,6],
    [6,7],
    [7,8]

])
#0=fail,1=pass
y=np.array([0,0,0,1,1,1])

#step2:create svm model
model=SVC(kernel='linear')

#step3:
#train the model
model.fit(X,y)
#4 hours study,5attendance
new_student=np.array([[4,5]])
prediction=model.predict(new_student)
print("Prediction(0=fail, 1=pass):",prediction[0])

#step5:create dicision boundary
xx,yy=np.meshgrid(np.linspace(0,9,100),np.linspace(0,9,100))
Z=model.predict(np.c_[xx.ravel(),yy.ravel()])
Z=Z.reshape(xx.shape)

#step6:plot graph
plt.figure()
plt.contourf(xx,yy,Z,alpha=0.3)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel("Study hours")
plt.ylabel("Attendance")
plt.title("SVM:Pass/Fail prediction")
plt.show()
```

```python
#example support vector vs svm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC


X=np.array([[0],[1],[2],[3],[4],[5]])
y=np.array([0,0,0,1,1,1])

model=SVC(kernel='linear')
model.fit(X,y)
new_email=[[3]]
prediction=model.predict(new_email)

if prediction[0]==1:
    print("This email is spam")
else:
    print("This email is not spam")
x_test=np.linspace(-1,6,100).reshape(-1,1)
y_pred=model.predict(x_test)

plt.figure()
plt.scatter(X,y)
plt.plot(x_test,y_pred)
plt.xlabel("Number of links in email")
plt.ylabel("Spam prediction (0=no, 1=yes)")
plt.title("SVM spam detection graph")
plt.show()


```

```python
#example for k neighbors
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

X=np.array(
    [
    [150,7], #Apple
    [170,8], #Apple
    [140,6], #Apple
    [130,6], #Apple
    [180,9], #Apple
    [200,10], #Orange
    [210,11], #Orange
    [190,9], #Orange

])

y=np.array([0,0,0,0,1,1,1,1])
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

new_fruit=np.array([[160,7]])
prediction=model.predict(new_fruit)
print("Prediction:","Apple" if prediction[0]==0 else "Orange")

x_min,x_max=X[:,0].min()-20,X[:,0].max() + 20
y_min,y_max=X[:,1].min()-2,X[:,1].max() + 2
xx,yy=np.meshgrid(np.arange(x_min, x_max, 1),np.arange(y_min,y_max,0.2))

z=model.predict(np.c_[xx.ravel(),yy.ravel()])
z=z.reshape(xx.shape)

plt.contourf(xx, yy, z, alpha=0.2)

for i, label in enumerate(y):
  if label==0:
    plt.scatter(X[i,0],X[i,1],marker='o',s=100,label="Apple"if i==0 else"")
  else:
    plt.scatter(X[i,0],X[i,1],marker='s',s=100,label="Orange"if i==4 else"")
plt.scatter(new_fruit[0,0],new_fruit[0,1],marker='*',s=200,label="New Fruit")

plt.xlabel("Weight")
plt.ylabel("Size")
plt.title("KNN Fruit Classification(Different style)")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X=np.array([[1],[2],[3],[4],[5],[6],[7],[8]])
y=np.array([0,0,0,1,1,1,1,1])

X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.25,random_state=42)

model=LogisticRegression()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)

print("Actual:",y_test)
print("Prediction:",y_pred)
print("Accuracy:",accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
X=np.array([
  [4,0], #basic
  [5,1], #basic
  [6,1], #basic
  [10,3], #premium
  [12,4], #premium
  [14,5], #premium
  [9,2], #premium
  [5,0], #basic
])
y=np.array([0,0,0,1,1,1,1,0])

model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
new_customer=np.array([[11,3]])
prediction=model.predict(new_customer)
print("predicted membership:","Premium"if prediction[0]==1 else "Basic")

for i in range(len(X)):
  if y[i]==0:
    plt.scatter(X[i][0],X[i][1],color='blue',label='Basic' if i==0 else "")
  else:
    plt.scatter(X[i][0],X[i][1],color='red',label='Premium' if i==3 else "")

plt.scatter(new_customer[0][0],new_customer[0][1],color='green',s=200,label='New Customer')
plt.xlabel("Monthly Visits")
plt.ylabel("Training Session")
plt.title("KNN gym membership classification")
plt.legend()
plt.show()



```

```python
#EXAMPLE FOR UNSUPERVISED LEARNING
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X=np.array([
    [22,2000],
    [25,2200],
    [23,2100],
    [45,8000],
    [50,8500],
    [48,8200]])
model=KMeans(n_clusters=2)
model.fit(X)
labels=model.labels_
centers=model.cluster_centers_
for i in range(len(X)):
  if labels[i]==0:
    plt.scatter(X[i][0],X[i][1],color='blue')
  else:
    plt.scatter(X[i][0],X[i][1],color='red')

plt.scatter(centers[:,0],centers[:,1],color='green',marker='X',s=200,label='centers')
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("K-Means Clustering")
plt.legend()
```

```python
#EXAMPLE FOR UNSUPERVISED LEARNING
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X=np.array([
    [40,45],
    [42,50],
    [38,43],
    [85,88],
    [90,92],
    [87,85],
    [60,65],
    [62,63]])
inertia=[]

for k in range(1,6):
  kmeans=KMeans(n_clusters=k,random_state=42)
  kmeans.fit(X)
  inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1,6),inertia,marker='o')
plt.xlabel("Number of clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()

#======================================================
#graph 2:cluster visulization
#=======================================================

model=KMeans(n_clusters=3,random_state=42)
model.fit(X)

labels=model.labels_
centers=model.cluster_centers_

colors=['blue','green','red']
for i in range(len(X)):
  plt.scatter(X[i][0],X[i][1],color=colors[labels[i]])

  plt.scatter(centers[:,0],centers[:,1],color='indigo',marker='X',s=200)
  plt.xlabel("Math marks")
  plt.ylabel("Science marks")
  plt.title("Student Clustering")
  plt.show()
  #======================================
  #graph3:cluster size bar chart

  unique,counts=np.unique(labels,return_counts=True)
  plt.figure()
  plt.bar(unique,counts)
  plt.xlabel("Cluster Number")
  plt.ylabel("Number of Students")
  plt.title("Cluster size Comparsion")
  plt.show()


```

```python
#EXAMPLE FOR UNSUPERVISED LEARNING
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
X=np.array([
    [40,45,42],
    [42,50,48],
    [38,43,40],
    [85,88,90],
    [90,92,95],
    [87,85,89],
    [60,65,63],
    [62,63,60]])
#train model
model=KMeans(n_clusters=3,random_state=42)
model.fit(X)
labels=model.labels_
centers=model.cluster_centers_

#heatmap of marks
plt.figure()
sns.heatmap(X,annot=True,cmap="coolwarm",xticklabels=["Math","Science","English"])
plt.title("Student marks heatmap")
plt.show()
```

```python
#3D clustering
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
for i in range(len(X)):
  ax.scatter(X[i][0],X[i][1],X[i][2],color=colors[labels[i]])
ax.scatter(centers[:,0],centers[:,1],centers[:,2],color='indigo',marker='X',s=200)
ax.set_xlabel("Math")
ax.set_ylabel("Science")
ax.set_zlabel("English")
ax.set_title("3D Student clustering")
plt.show()
```
